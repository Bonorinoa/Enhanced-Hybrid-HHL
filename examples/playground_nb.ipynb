```python
from qiskit.circuit.library import HamiltonianGate
from qiskit_ibm_provider import IBMProvider
from qiskit import transpile, QuantumCircuit
from qiskit.quantum_info import random_hermitian
import numpy as np
```

```python
provider = IBMProvider()
backend = provider.get_backend('ibm_torino')
```

```python
matrix = random_hermitian(4)
matrix
```

```python
matrix = np.diag([-1,-0.6, 0.3, 0.4, 0.7,1])
```

```python
gate = HamiltonianGate(matrix, 2*np.pi)

circ = QuantumCircuit(gate.num_qubits)
circ.append(gate, circ.qubits)
```

```python
transp = transpile(circ, backend)
transp.depth()
```

```python
transp.draw()
```
